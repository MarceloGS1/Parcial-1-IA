```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")
# Mostrar las primeras filas del csv
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

clase_modo = df['mode']
label_encoder = LabelEncoder()
clase_modo_codificado = label_encoder.fit_transform(clase_modo)
print("0 = Major, 1 = Minor")
print(clase_modo_codificado)

```

```python
import numpy as np
#numpy lo uso para convertir mi lista de datos

artist_count = df['artist_count']  #obtengo la columna del csv
# Para convertir la serie en una lista.
lista_datos = artist_count.tolist()

X = np.array(lista_datos).reshape(-1, 1)
y = clase_modo_codificado

# Normalización de los datos (Min-Max)
def normalizar(X):
    X_normalizado = []
    for col in range(len(X[0])):
        column_values = [row[col] for row in X]
        min_val = min(column_values)
        max_val = max(column_values)
        normalized_col = [(val - min_val) / (max_val - min_val) for val in column_values]
        X_normalizado.append(normalized_col)

    # Transponer para obtener filas
    return list(map(list, zip(*X_normalizado)))

# Función para penalización L1 (Lasso)
def lasso_penalty(X, y, alpha):
    X_normalizado = normalizar(X)
    weights = [0.0] * len(X_normalizado[0])
    learning_rate = 0.01
    iterations = 1000

    for _ in range(iterations):
        for i in range(len(y)):
            prediction = sum(weights[j] * X_normalizado[i][j] for j in range(len(weights)))
            error = prediction - y[i]
            for j in range(len(weights)):
                weights[j] -= learning_rate * (error * X_normalizado[i][j] + alpha * (1 if weights[j] > 0 else -1))

    return weights

# Función para penalización L2 (Ridge)
def ridge_penalty(X, y, alpha):
    X_normalizado = normalizar(X)
    weights = [0.0] * len(X_normalizado[0])
    learning_rate = 0.01
    iterations = 1000

    for _ in range(iterations):
        for i in range(len(y)):
            prediction = sum(weights[j] * X_normalizado[i][j] for j in range(len(weights)))
            error = prediction - y[i]
            for j in range(len(weights)):
                weights[j] -= learning_rate * (error * X_normalizado[i][j] + alpha * weights[j])

    return weights

# Ejemplo de uso
alpha = 0.1
lasso_weights = lasso_penalty(X, y, alpha)
ridge_weights = ridge_penalty(X, y, alpha)

print("Lasso L1:", lasso_weights)
print("Ridge L2:", ridge_weights)
```
