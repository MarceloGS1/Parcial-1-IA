```python
#OnehotEncoder
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")

key = df['key']  #obtengo la columna del csv
# Para convertir la serie en una lista.
lista_datos = key.tolist()

# Datos categóricos
data = pd.DataFrame({'key': ["'A#'" ,"'A'" ,"'B'" ,"'C#'" ,"'D#'", "'D'", "'E'", "'F#'", "'F'", "'G#'", "'G'", "'none'"]})

# Aplicar One-Hot Encoding
onehot_encoded = pd.get_dummies(data, columns=['key'], prefix='', prefix_sep='')
onehot_encoded = onehot_encoded.astype(int)
# Mostrar los resultados
print("Original:\n", data)
print("\nOne-Hot Encoding:\n", onehot_encoded)


```

```python
#LabelEncoder
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")

key = df['key']  #obtengo la columna del csv
# Para convertir la serie en una lista.
lista_datos = key.tolist()

# Crear el codificador
label_encoder = LabelEncoder()

# Aplicar label encoding
encoded_labels = label_encoder.fit_transform(lista_datos)

# Mostrar los resultados
print("Categorico Original:")
print(lista_datos)
print("Etiquetado:")
print(encoded_labels)
print("Todas las posibles etiquetas")
print(label_encoder.classes_)
```

```python
#Discretizar
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")
artist_count = df['artist_count']  #obtengo la columna del csv
lista_datos = artist_count.tolist() # Para convertir la serie en una lista.
# Crear un array 1D
lista_datos = np.array(lista_datos)

# Definir los límites de los bins
bins = [0,1,2,3]

# Discretizar el array
lista_discretizado = np.digitize(lista_datos, bins)

# Mostrar resultados
print("Array Original\n:", lista_datos)
print("Array Discretizado:\n", lista_discretizado)
```

```python
#Normalizar usando librerias

from sklearn.preprocessing import MinMaxScaler
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")
artist_count = df['artist_count']  #obtengo la columna del csv
lista_datos = artist_count.tolist() # Para convertir la serie en una lista.

array_1d = np.array(lista_datos).reshape(-1, 1)

# Crear el escalador Min-Max
scaler = MinMaxScaler()
# Normalizar
lista_normalizada = scaler.fit_transform(array_1d)
# Mostrar resultados
#flatten(aplana los datos para que vuelvan a estar en 1D o sea una lista)
print("Array Original:\n", array_1d.flatten())
print("\nArray Normalizado:\n", lista_normalizada.flatten())
```
