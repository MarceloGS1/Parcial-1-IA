```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")

speechiness_ = df['speechiness_%']  #obtengo la columna del csv
# Para convertir la serie en una lista.
lista_datos = speechiness_.tolist()

# Calcular la media, mediana y moda
media = np.mean(speechiness_)
mediana = np.median(speechiness_)
moda = stats.mode(speechiness_)[0]

# Crear el diagrama de caja y bigote
plt.figure(figsize=(5, 6))
sns.boxplot(data=speechiness_, color='lightblue')

# Añadir líneas para la media y moda
plt.axhline(media, color='red', linestyle='--', label='Media')
plt.axhline(mediana, color='green', linestyle='-', label='Mediana')
plt.axhline(moda, color='orange', linestyle=':', label='Moda')

# Añadir leyenda y título
plt.legend()
plt.title('Diagrama de Caja y Bigote con Media, Mediana y Moda \nde las palabras de una cancion')
plt.ylabel('Valores')

# Mostrar gráfico
plt.show()

print("Explicacion")
print("Por un lado vemos que existen datos atipicos muy alejados del diagrama caja-bigote \ntambien vemos que la media se acerca al Q1 cuartil 1 \nextremo de arriba de la caja y la mode se acerca al Q3")
```
