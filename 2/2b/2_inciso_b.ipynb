```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv("/content/drive/MyDrive/Dataset_Spotify/Spotify_Most_Streamed_Songs.csv")

bailable = df['danceability_%']
palabras = df['speechiness_%']
instrumentos = df['instrumentalness_%']

clase_modo = df['mode']
label_encoder = LabelEncoder()
clase_modo_codificado = label_encoder.fit_transform(clase_modo)

# Crear el scatter plot
plt.figure(figsize=(8,6))

#CASO 1 columna Bailable vs Palabras
plt.scatter(bailable, palabras, c=clase_modo_codificado, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel('bailable')
plt.ylabel('palabras')
plt.title('Scatter Plot de bailable vs palabras con modo como Categoría')
plt.colorbar(label='Tonalidad (0: Major, 1: Minor)')
plt.show()

#CASO 2 columna Bailable vs instrumentos
plt.scatter(bailable, instrumentos, c=clase_modo_codificado, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel('bailable')
plt.ylabel('instrumentos')
plt.title('Scatter Plot de bailable vs instrumentos con modo como Categoría')
plt.colorbar(label='Tonalidad (0: Major, 1: Minor)')
plt.show()

#CASO 3 columna instrumentos vs palabras
plt.scatter(instrumentos, palabras, c=clase_modo_codificado, cmap='coolwarm', edgecolor='k', s=50)
plt.xlabel('instrumentos')
plt.ylabel('palabras')
plt.title('Scatter Plot de instrumentos vs palabras con modo como Categoría')
plt.colorbar(label='Tonalidad (0: Major, 1: Minor)')
plt.show()


print("------CARACTERISTICAS danceability_%--------")
print("En esta columna podemos analizar que tan bailable es una cancion,")
print("el valor esta en un rango porcentual por que los valores van del 0-100")
print("Si analizamos la grafica de dispercion de bailabilidad vs palabras ")
print("podemos ver que en la mayoria de las canciones existe una relacion entre \nvariables, pero como el 25% de los datos no tienen relacion\n  ")

print("------CARACTERISTICAS speechiness_%--------")
print("Esta columna nos habla de la cantidad de palabras que existen en una cancion \nigualmente lo hace con porcentajes, si analizamos \nla grafica de dispercion de palabras vs instrumentos, podemos ver que mientras \nmenos se escuchen los instrumentos en una cancion hay mas palabras, \npor lo que si existe una relacion entre estos dos\n")

print("------CARACTERISTICAS instrumentalness_%--------")
print("Aqui nos habla de la presencia de instrumentos en una cancion, en el \nDataset se puede observar que la mayoria de canciones populares carece de instrumentos \no solo usan programas para generar la cancion. \ny si analizamos la grafica de dispercion de bailable vs instrumentos, vemos que la mayoria de datos estan relacionados")
```
